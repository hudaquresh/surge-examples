# NHC Forecast Verification 

```python
def read_file(filename, year):
    # Read file, remove header
    lines = [line.rstrip('\r\n') for line in open(filename)]
    
    # Transform header keys into a list of words
    header_keys = lines[7]
    header_keys = header_keys.split()

    # Filter in data removing the header 
    no_header = lines[8:]
#     print(len(header_keys))
#     print(len(lines[9].split()))
    
#     print(header_keys)
#     print(lines[9].split())

    # filter by year
    year = str(year)
    data_array = []
    no_header[0:4]
    test_line = True 
    for line in no_header:
        if test_line == True: 
            if int(line[6:10]) >= int(year):
#                 print(line)
#                 print(len(line.split()))
                data_array.append(line.split())
#                 print(len(line))
        else: 
            if int(line[6:10]) >= int(year):
                print(line)
                templine = []
                buffer = ""
                for i in range(len(line)):
                    if line[i] != ' ':
                        buffer += line[i]
                    else:
                        if buffer != "":
                            templine.append(buffer)
                            buffer = ""
                print(templine)
                data_array.append(templine)
    # Sort
    storms = []
    storm_name = data_array[0][1]
    tempstorm = []

    for line in data_array:
        if line[1] == storm_name:
            tempstorm.append(line)
        else:
            storms.append(tempstorm)
            storm_name = line[1]
            tempstorm = [line]
    return storms, header_keys 
```

In the above we have a method that allows us to filter all the storms into it's own list. We even return the header keys. This will allow us to transform the data into a dictionary. That will then be transformed into a pandas dataframe (Don't judge me this is the only way I know how to do this) 

Below we use the header keys we collected to create a dictionary using those keys. Then we run through the storms to get those keys assigned to their appropriate values. 
Once this is complete we can easily transform into a pandas dataframe.

```python
storms, header_keys = read_file('1970-present_OFCL_v_BCD5_ind_ATL_TI_errors_noTDs.txt', 2000)
dict_data = {}
for key in header_keys: 
    dict_data[key] = []
# print(len(storms))
for i in range(0, len(storms)): 
    for j in range(0, len(storms[i])): 
        for n in range(0,54): 
            if n == 0 or n == 1: 
                dict_data[header_keys[n]].append(storms[i][j][n])
#             elif n == 53: 
#                 dict_data[header_keys[n]].append(0)
            else: 
                dict_data[header_keys[n]].append(float(storms[i][j][n]))

# print(header_keys)
```

## Pandas Data Frame of the Data Set

We transform the dictionary easily into a pandas data frame. Accordingly we will create plots of the 000hI01 versus the wind speed to see how easily we can manipulate the data. 

```python
pandas_error_data = pandas.DataFrame(dict_data)
# pandas_error_data

# Plot pandas data using the example provided from the resource below
# Here we needed matplotlib to determine how to properly use the maker
# Link: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html

fig = plt.figure
df = pandas_error_data.copy()
ws = numpy.zeros((1, len(df)))
y = numpy.zeros((1, len(df)))

for i in range(0, ws.shape[1]): 
    ws[0,i] = df['WS'][i]
    if df['000hI01'][i] < -800: 
        y[0,i] = 0 
    else: 
        y[0,i] = df['000hI01'][i]
    
# pandas_error_data.plot(x='WS', y='000hI01', color='green', marker='o')
plt.plot(ws, y, 'b*')
plt.title('Plot of Intensity Errors at t=0 versus wind speed')
plt.show() 
```
